# Define $G_{OL}(s)$ for concentration model:

```python
import sympy as sy
from sympy import exp, sin, cos, latex

j = sy.I # imaginary number
K_c, tau_I, s , omega, K_p1, tau_p1, theta_C, K_IP, K_m, K_v = sy.symbols('K_c tau_I s omega K_p1 tau_p1 theta_C K_IP K_m K_v')

G_c1 = K_c*(tau_I*s + 1)/(tau_I*s)
G_c1
```

```python
print(latex(G_c1))
```

```python
G_p1 = K_p1*exp(-theta_C*s)/(tau_p1*s + 1)
G_p1
```

```python
print(latex(G_p1))
```

Open-loop transfer function is obtained from all elements in the feedback loop, if the closed loop is opened up just before the comparator.
$$
G_{OL} = G_c G_{IP} G_v G_{p1} G_m
$$
Neglect valve and sensor dynamics:
$$
G_v = K_v
$$

$$
G_{IP} = K_{IP}
$$

$$
G_m = K_m
$$

$$
\therefore G_{OL} = K_{IP} K_m K_v G_c G_{p1}
$$

```python
G_OL = K_IP*K_m*K_v*G_c1*G_p1
G_OL
```

```python
print(latex(G_OL))
```

Transform to frequency domain: <br>
Obtain expression for $G_{OL}(j \omega)$

```python
G_OLjw = G_OL.subs(s, j*omega)
G_OLjw
```

```python
print(latex(G_OLjw))
```

The dead-time term can be replaced with the Euler identity:
$$
e^{-j \omega \theta_C} = \cos{\omega \theta_C} -j \sin{\omega \theta_C}
$$

```python
G_OLjw = j*K_IP*K_c*K_m*K_p1*K_v*(j*omega*tau_I + 1)*(cos(omega*theta_C) - j*sin(omega*theta_C))/(omega*tau_I*(j*omega*tau_p1 + 1))
G_OLjw.expand()
```

```python
print(latex(G_OLjw))
```

```python
G_OLjw = G_OLjw*1/(omega*tau_I - j*omega**2*tau_I*tau_p1)
G_OLjw
```

This can be written into the following form if the denominator is simplified:
$\omega \tau_I$ can be factored out of the second bracket in the denominator to give:
$$
G_{OL}(j \omega) = \frac{j K_{IP} K_c K_m K_{p1} K_v(\cos(\omega \theta_C) - j \sin(\omega \theta_C))(1 + j \omega \tau_I)}{\omega^2 \tau_I^2(1 + j \omega \tau_{p1})(1 - j \omega \tau_{p1})}
$$
The denominator is transformed, using the result that $j^2 = -1$:
$$
G_{OL}(j \omega) = \frac{j K_{IP} K_c K_m K_{p1} K_v(\cos(\omega \theta_C) - j \sin(\omega \theta_C))(1 + j \omega \tau_I)}{\omega^2 \tau_I^2(1 - j^2 \omega^2 \tau_{p1}^2)}
$$
$$
G_{OL}(j \omega) = \frac{j K_{IP} K_c K_m K_{p1} K_v(\cos(\omega \theta_C) - j \sin(\omega \theta_C))(1 + j \omega \tau_I)}{\omega^2 \tau_I^2(1 + \omega^2 \tau_{p1}^2)}
$$

```python
G_OLjw = j*K_IP*K_c*K_m*K_p1*K_v*(cos(omega*theta_C) - j*sin(omega*theta_C))*(1 + j*omega*tau_I)/(omega**2*tau_I**2*(1 + omega**2*tau_p1**2))
G_OLjw
```

This expression also has to be multiplied by the complex conjugate in the numerator to finish rationalisation:

```python
G_OLjw*(omega*tau_I - j*omega**2*tau_I*tau_p1)
```

```python
G_OLjw.expand()
```

The product of the gains is commonly referred to as $K_{OL}$:
$$
K_{OL} = K_{IP} K_c K_m K_{p1} K_v
$$

The real and imaginary parts are obtained from $G_{OL}(j \omega)$ :
$$
\text{Re} \left[G_{OL}(j \omega)\right] = \frac{K_{OL} \left[\sin{(\omega \theta_C)} - \omega \tau_I \cos{(\omega \theta_C)}\right]}{\omega^2 \tau_I^2 + \omega^4 \tau_I^2 \tau_{p1}^2}
$$

$$
\text{Im} \left[G_{OL}(j \omega)\right] = \frac{K_{OL} \left[\cos{(\omega \theta_C)} + \omega \tau_I \sin{(\omega \theta_C)}\right]}{\omega^2 \tau_I^2 + \omega^4 \tau_I^2 \tau_{p1}^2}
$$

The amplitude ratio (AR) and phase angle can then be calculated:

$$
AR_{OL} = \Big \vert G_{OL}(j \omega) \Big \vert = \sqrt{\text{Re} \left[G_{OL}(j \omega)\right]^2 + \text{Im} \left[G_{OL}(j \omega)\right]^2}
$$

$$
\phi = \angle G_{OL}(j \omega) = \tan^{-1}{\left[\frac{\text{Im} \left[G_{OL}(j \omega)\right]}{\text{Re} \left[G_{OL}(j \omega)\right]}\right]}
$$

The gold loading is also controlled with PI, so it has the same expression for $G_{OL}(j \omega)$.

# Temperature loop:

```python
tau_D, K_p3, tau_p3, theta_T, K_OL3 = sy.symbols('tau_D K_p3 tau_p3 theta_T K_OL3')

G_c3 = K_c*(1 + tau_I*s + tau_I*tau_D*s**2)/(tau_I*s)
G_c3
```

```python
print(latex(G_c3))
```

```python
G_p3 = K_p3*exp(-theta_T*s)/(tau_p3*s + 1)
G_p3
```

```python
print(latex(G_p3))
```

```python
G_OL3 = K_IP*K_m*K_v*G_c3*G_p3
G_OL3
```

```python
print(latex(G_OL3))
```

Simplify:
$$
K_{OL3} = K_{IP} K_c K_m K_{p3} K_v
$$

```python
G_OL3 = K_OL3*(s**2*tau_D*tau_I + s*tau_I + 1)*exp(-s*theta_T)/(s*tau_I*(s*tau_p3 + 1))
G_OL3
```

```python
G_OLjw3 = G_OL3.subs(s, j*omega)
G_OLjw3
```

Substitute Euler's identity:

```python
G_OLjw3 = j*K_OL3*(-omega**2*tau_D*tau_I + j*omega*tau_I + 1)*(cos(omega*theta_T) - j*sin(omega*theta_T))/(omega*tau_I*(j*omega*tau_p3 + 1))
G_OLjw3
```

Multiply by complex conjugate, first in denominator:

```python
conj = 1 - j*omega*tau_p3
G_OLjw3 = G_OLjw3*1/conj
G_OLjw3
```

```python
print(latex(G_OLjw3))
```

This expression can simplify to:
$$
\frac{j K_{OL3} \left(- j \sin{\left(\omega \theta_{T} \right)} + \cos{\left(\omega \theta_{T} \right)}\right) \left(- \omega^{2} \tau_{D} \tau_{I} + j \omega \tau_{I} + 1\right)}{\omega \tau_{I} \left(1 - j^2 \omega^2 \tau_{p3}^2 \right)}
$$
Using $j^2 = -1$:
$$
\frac{j K_{OL3} \left(- j \sin{\left(\omega \theta_{T} \right)} + \cos{\left(\omega \theta_{T} \right)}\right) \left(- \omega^{2} \tau_{D} \tau_{I} + j \omega \tau_{I} + 1\right)}{\omega \tau_{I} \left(1 + \omega^2 \tau_{p3}^2 \right)}
$$

```python
G_OLjw3 = j*K_OL3*(-omega**2*tau_D*tau_I + j*omega*tau_I + 1)*(cos(omega*theta_T) - j*sin(omega*theta_T))/(omega*tau_I*(1 + omega**2*tau_p3**2))
G_OLjw3
```

Now multiply conjugate in the numerator:

```python
G_OLjw3*conj
```

```python
G_OLjw3.expand()
```

```python
print(latex(G_OLjw3.expand()))
```

Write real and imaginary parts:
$$
\text{Re} \left[G_{OL}(j \omega)\right] = \frac{K_{OL3} \left[\sin{(\omega \theta_T)} + \omega^2 \tau_D \tau_I \sin{(\omega \theta_T)} - \omega \tau_I \cos{(\omega \theta_T)}\right]}{\omega \tau_{I} + \omega^{3} \tau_{I} \tau_{p3}^{2}}
$$

$$
\text{Im} \left[G_{OL}(j \omega)\right] = \frac{K_{OL3} \left[\cos{(\omega \theta_T)} - \omega^2 \tau_D \tau_I \cos{(\omega \theta_T)} + \omega \tau_I \sin{(\omega \theta_T)}\right]}{\omega \tau_{I} + \omega^{3} \tau_{I} \tau_{p3}^{2}}
$$
